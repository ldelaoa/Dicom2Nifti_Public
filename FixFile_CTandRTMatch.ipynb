```python
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('//zkh/appdata/RTDicom/Projectline - Modelling lung cancer outcomes [panama code]/Users/Luis/CSVs/CTandRTMatch.csv')

# Create a new DataFrame to store the modified data
new_df = pd.DataFrame(columns=df.columns)

# Iterate through each row in the original DataFrame
for index, row in df.iterrows():
    # Split the paths using a comma
    pathsTot = row['Tot'].split(',')
    pathsTumor = row['Tumor'].split(',')
    pathsLN = row['LymphNodes'].split(',')
    pathsCT = row['CT'].split(',')
    
    # If there is only one path, keep the row as it is in the new DataFrame
    if len(pathsTot) == 1:
        new_df = pd.concat([new_df, pd.DataFrame([row])], ignore_index=True)
    else:
        print("Repeated",index)
        count=0
        for path in pathsTot:
            # Create a copy of the original row
            new_row = row.copy()
            # Update the path in the new row
            new_row['Tot'] = path.strip()
            
            if count<len(pathsCT):
                new_row['CT'] = pathsCT[count].strip()
            if count<len(pathsTumor):
                new_row['Tumor'] = pathsTumor[count].strip()
            if count<len(pathsLN):
                new_row['LymphNodes'] = pathsLN[count].strip()            
            
            # Concatenate the modified row to the new DataFrame
            new_df = pd.concat([new_df, pd.DataFrame([new_row])], ignore_index=True)

            count+=1
            
# Remove "[" and "]" from all columns in the new row
new_df['CT'] = new_df['CT'].str.replace('[\[\]]', '', regex=True)
new_df['Tot'] = new_df['Tot'].str.replace('[\[\]]', '', regex=True)
new_df['Tumor'] = new_df['Tumor'].str.replace('[\[\]]', '', regex=True)
new_df['LymphNodes'] = new_df['LymphNodes'].str.replace('[\[\]]', '', regex=True)

# Save the modified DataFrame to a new CSV file
new_df.to_csv('//zkh/appdata/RTDicom/Projectline - Modelling lung cancer outcomes [panama code]/Users/Luis/CSVs/CTandRTMatch_modified_file.csv', index=False)
```
