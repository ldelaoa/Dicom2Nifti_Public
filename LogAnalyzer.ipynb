```python
import re

# Specify the path to your log file
log_file_path = '//zkh/appdata/RTDicom/Projectline_modelling_lung_cancer/Users/Luis/Logs/LogUnEdited.csv'

# Specify the output file path
output_file_path = '//zkh/appdata/RTDicom/Projectline_modelling_lung_cancer/Users/Luis/Logs/LogEditedED.csv'

with open(log_file_path, 'r', encoding='utf-8-sig') as log_file:

    lines = log_file.readlines()

# Initialize variables
current_px_id = ""
output_data = []
```

```python

# Iterate through the lines in the log file
pxCount=0
listToSave = []
for line in lines:
    if line.startswith("Px "):
        if pxCount>0:
            print(current_px_id,missingLabels_count,rtStructInvalid_count,rtNotLinked_count,NoDCMfound,NoRT_count,NoCT_count)
            listToSave.append([current_px_id,missingLabels_count,rtStructInvalid_count,rtNotLinked_count,NoDCMfound,NoRT_count,NoCT_count])
        current_px_id = line.strip()
        missingLabels_count,rtStructInvalid_count,rtNotLinked_count,NoDCMfound,NoRT_count,NoCT_count = (0,0,0,0,0,0)
        pxCount+=1
    if "No labels Matched for" in line:
        missingLabels_count+=1
    if "Please check that the existing RTStruct is valid" in line:
        rtStructInvalid_count+=1
    if "Loaded RTStruct references image(s) that are not contained in input series data. Problematic image has SOP Instance Id" in line:
        rtNotLinked_count+=1
    if "No CT" in line:
        NoCT_count+=1
    if "No DICOM Images found in input path" in line:
        NoDCMfound+=1
    if "RTs 0" in line:
        NoRT_count+=1
    
```

```python
import csv
with open(output_file_path, 'w', newline='') as output_file:
    csv_writer = csv.writer(output_file)

    # Write the rows to the CSV file
    csv_writer.writerows(listToSave)
```
