```python
import os
import pandas as pd
import shutil
root_p = "//zkh/appdata/RTDicom/Projectline - Modelling lung cancer outcomes [panama code]/Users/Luis/"

# Specify the root directories for Folder A and Folder B
folder_a_root = root_p+'CTRT_NiftiFiles_MatchID_PETIDs/'
folder_b_root = root_p+'PET_NiftiFiles_MatchID'
```

```python

def get_file_paths(directory):
    acct_path = None
    pet_path = None
    ct_path = None
    rtstruct_path = None
    for filename in os.listdir(directory):
        full_path = os.path.join(directory, filename)
        if os.path.isfile(full_path):
            if "AC_CT_Body" in filename or "CT van PET" in filename or "CT LD" in filename or "AC CT" in filename or "AC  CT" in filename:
                acct_path = full_path   
            elif 'PET' in filename  or "Earl" in filename:
                pet_path = full_path
            elif "%" in filename or "thorax" in filename or "ave" in filename or "mip" in filename or ("ct" in filename and not "rtstruct" in filename):
                ct_path = full_path
            elif 'ITV'in filename  or "IGTV" in filename  or "IgTV" in filename:
                rtstruct_path = full_path
    return acct_path, pet_path, ct_path, rtstruct_path

# Get all unique IDs from both folders
all_ids = set(os.listdir(folder_a_root)).union(set(os.listdir(folder_b_root)))

# Initialize lists to store data
ids = []
acct_paths = []
pet_paths = []
ct_paths = []
rtstruct_paths = []

# Iterate through all IDs
for id_folder in all_ids:
    id_folder_a_path = os.path.join(folder_a_root, id_folder)
    id_folder_b_path = os.path.join(folder_b_root, id_folder)
    
    # Check if the ID folder exists in Folder A and Folder B
    if os.path.isdir(id_folder_a_path) and os.path.isdir(id_folder_b_path):
        ids.append(id_folder)
        #PET Folders
        acct_path, pet_path, _, _ = get_file_paths(id_folder_b_path)
        acct_paths.append(acct_path)
        pet_paths.append(pet_path)
        #CT Folders
        _, _, ct_path, rtstruct_path = get_file_paths(id_folder_a_path)
        ct_paths.append(ct_path)
        rtstruct_paths.append(rtstruct_path)

# Create a DataFrame
data = {'ID': ids, 'ACCT': acct_paths, 'PET': pet_paths, 'CT': ct_paths, 'RTStruct': rtstruct_paths}

df = pd.DataFrame(data)

df
```

```python
df_cleaned = df.dropna()

df_cleaned
```

```python
# Save the DataFrame to a CSV file
df_cleaned.to_csv(root_p+'/CSVs/CompleteNii_ACCT_PET_CT_RTStruct.csv', index=False)
```

```python
destination_folder = root_p+'CT_RT_PET_ACCT_NiiToHabrok/' 
source_folder = root_p
# Iterate through rows in the DataFrame
for index, row in df_cleaned.iterrows():
    count=0
    # Iterate through columns and copy files if the path is not None
    for column in df_cleaned.columns:
        cell_info = row[column]
        if count == 0 :
            destination_folder_px = os.path.join(destination_folder,cell_info)
        else:
            temp_cell_info = cell_info.replace('\\', '/')
            file_name = temp_cell_info.split('/')[-3:]
            
            source_folder_px_final = os.path.join(source_folder,file_name[0],file_name[1],file_name[2])
            destination_folder_px_final = os.path.join(destination_folder_px,file_name[2])
            if not os.path.exists(destination_folder_px):
                os.mkdir(destination_folder_px)
            shutil.copy(source_folder_px_final, destination_folder_px_final)
            
        count+=1
    
            
```
