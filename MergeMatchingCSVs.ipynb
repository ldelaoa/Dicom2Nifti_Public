```python
import pandas as pd

root_p = "//zkh/appdata/RTDicom/Projectline - Modelling lung cancer outcomes [panama code]/Users/Luis/CSVs/"
file_CT = pd.read_csv(root_p+'CTandRTMatch_modified_file.csv')
file2_PET = pd.read_csv(root_p+'PET_DICOM_Paths.csv')

#Header CT: Px	CT	Tot	Tumor	LymphNodes
#Header PET: Px	ACCTBody	PET

merged_df = pd.merge(file_CT, file2_PET, on='Px', how='inner')

# Filter rows where all paths are non-null
filtered_df = merged_df.dropna(subset=['CT', 'Tot', 'PET'])

# Create a new dataframe with the desired columns
result_df = filtered_df[['Px', 'CT', 'Tot', 'PET', 'ACCTBody']]

result_df
```

```python
# Save the result to a new CSV file
result_df.to_csv(root_p+'MergedCT_RT_PET_ACCT.csv', index=False)
```
