```python
import pandas as pd
import numpy as np
```

```python
import yaml
with open('config.yaml', 'r') as stream:
    config = yaml.safe_load(stream)
    complete_collection_scans = config['complete_collection_of_scans_per_patient']
    
df = pd.read_excel(complete_collection_scans, index_col=0)
df.head()
```

Checking for empty cells

```python
df.isna().sum().sum()
```

```python
df.index.nunique()
```

```python
print(len(df[df.isna().any(axis=1)]))
df[df.isna().any(axis=1)]
```

```python
df.fillna('no_name_found', inplace=True)
```

Changing date into datetime format

```python
df4d = df[df['series_description'].str.contains('4D|4d|\d{1,2}%')]
```

```python
# filling in the missing dates with the dates the other scans have for this patient
no_date_scans = df[~df['study_date'].str.contains('\d{4}-\d{2}-\d{2}')]
for index in no_date_scans.index:

    value = df.loc[index, 'study_date'].unique()[-1]
    df.loc[index, 'study_date'] = value
```

```python
df['study_date'] = pd.to_datetime(df['study_date'])
```

```python
df['study_date'] = df['study_date'].dt.strftime('%Y')
```

```python
df4d = df[df['series_description'].str.contains('4D|4d|\d{1,2}%')]
x = df4d.reset_index()[['p_id', 'study_date', 'study_description']].drop_duplicates()
x = x.groupby(['study_date', 'study_description']).size().unstack()
```

```python
x
```

```python
import hvplot.pandas
import holoviews as hv
x.fillna(0, inplace=True)
fig1 = x.hvplot(
    kind='bar', 
    stacked=True, 
    rot=0,
    width = 1200, height = 600,
    xlabel = 'Year',
    ylabel = 'Number of scans', 
    title='4D serie names over time',
    ).opts(
        legend_offset=(10, 2),
        )

hv.save(fig1, '4d_series_over_time.html')
fig1
```
