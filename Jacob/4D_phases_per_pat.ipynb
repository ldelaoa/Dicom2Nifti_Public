```python
import pandas as pd
import numpy as np
import yaml
with open('config.yaml', 'r') as stream:
    config = yaml.safe_load(stream)
    clean_scancollection = config['clean_collection_of_scans_per_patient']
    
df = pd.read_csv(clean_scancollection, index_col=0)
```

```python
# excluding all non 4D-CT scans from this df:
df = df[df['modality'] == 'CT']

# looking only for percentage description names, because including 4D returns some average scans that do not include the words Averages or MIP
# to check you can use this code: 
        # df[df['series_description'].str.contains('4D')&
        #    ~(df['series_description'].str.contains('\d{1,2}%|Ave|MIP'))]
        
df_4d = df[(df['series_description'].str.contains('\d{1,2}%')) &
           ~(df['series_description'].str.contains('Ave|MIP'))]
```
